# Fake LinkedIn Profile Detection

## 1. Data Loading

## Preprocessing
- Created binary label (0 → Legit, 1 → Fake)
- Concatenated About, Experiences, Educations, Skills into `full_text`
- Applied basic text cleaning (lowercasing, whitespace normalization)

```python
# Binary label
df["binary_label"] = df["Label"].apply(lambda x: 0 if x == 0 else 1)

# Concatenate text sections
df["full_text"] = (
    df["About"].astype(str) + " " +
    df["Experiences"].astype(str) + " " +
    df["Educations"].astype(str) + " " +
    df["Skills"].astype(str)
)

# Light text cleaning
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    return text

df["full_text"] = df["full_text"].apply(clean_text)
```

## Train-Test Split
- Stratified split to maintain class balance
- 80% training, 20% testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df["full_text"],
    df["binary_label"],
    test_size=0.2,
    stratify=df["binary_label"],
    random_state=42
)
```

## Feature Extraction
- TF-IDF vectorization of combined text
- Top 5000 words considered

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000, stop_words="english")
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

## Logistic Regression Model
- Baseline classifier for fake profile detection

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_tfidf, y_train)
predictions = model.predict(X_test_tfidf)
```

## Evaluation
- Classification report with precision, recall, F1-score
- Confusion matrix visualization
- Baseline text-only model achieved ~97% accuracy

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print(classification_report(y_test, predictions))

cm = confusion_matrix(y_test, predictions)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## Conclusion
- Text-based TF-IDF + Logistic Regression achieved ~97% accuracy
- Successfully distinguishes legitimate and fake LinkedIn profiles
- Next step: Hybrid model combining structured features

```python
import joblib

joblib.dump(model, "../model/fake_profile_model.pkl")
joblib.dump(vectorizer, "../model/vectorizer.pkl")
```

## 7. Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train_tfidf, y_train)

rf_predictions = rf_model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, rf_predictions))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, rf_predictions)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest Confusion Matrix")
plt.show()
```

## Model Comparison

Model 1: Logistic Regression  
Model 2: Random Forest

```python
from sklearn.metrics import accuracy_score

log_acc = accuracy_score(y_test, predictions)
rf_acc = accuracy_score(y_test, rf_predictions)

print("Logistic Regression Accuracy:", log_acc)
print("Random Forest Accuracy:", rf_acc)
```
